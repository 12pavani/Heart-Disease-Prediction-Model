# importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# data collection and processing

```python
# loading the csv data to a Pandas DataFrame
# https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset?resource=download
```

```python
heart_data = pd.read_csv('heart.csv')
```

```python
# splitting the feature and target
# here features are other column and target are target column
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the Data into Training data and Test data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model training
# Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

# Accuracy Score

```python
# accuracy on training data
# more than 75% accuracy is good 
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
# the accuracy of the traing data and test data should be almost similar to each other, if accuracy of trainging data and test data is way far from each other and accuracy score which is score on test data is very small, then model is over fitted
# No over fit issue here
```

# Building a predictive system

```python
input_data = (43,0,0,132,341,1,0,136,1,3,1,0,3)
```

```python
# change the input data to a numpy array
# asarray will chane the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)
```

```python
# reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
```

```python
prediction = model.predict(input_data_reshaped)
print(prediction)
```
